##Problem 2: Plot From Description

```python
# Log Fibonacci Numbers Plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# recursive fibonacci function from rec.py
def fib(n, verbose = False):
    "nth Fibonacci number"
    if verbose:
        print("fib called with n = {}".format(n))

    if n <= 1:
        return n  #fib(0) = 0, fib(1) = 1
    
    # n is greater than 1, use recursive call
    
    return fib(n-1, verbose) + fib(n-2, verbose)
```

```python
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize = (6,6), dpi = 120)
n = np.linspace(0, 20, 21)
# inputs

fib_n = np.array([fib(x) for x in n])
# the first n fibonacci sequence numbers

ax.scatter(n, np.log(fib_n), s = 6, c = 'blue')

c = np.log( (1 + np.sqrt(5)) / 2)
y = c * n

ax.plot(n, y, linewidth = 0.9, color = 'red')

ax.set_xlabel('n')
ax.set_ylabel('y')
plt.title('Log Fibonacci Numbers')

plt.show()

```
